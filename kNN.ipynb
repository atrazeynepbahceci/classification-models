```python
import seaborn
```

```python
df=seaborn.load_dataset("titanic")
```

```python
import pandas as pd
```

```python
df=df.drop(["class","adult_male","who","embark_town","alive","alone"], axis=1) 
#removes columns that repeat each other
```

```python
df = df.replace(['male'],'1') 
df=df.replace(['female'],"0")
#creates a dummy variable for gender
```

```python
df=pd.get_dummies(df, columns=["embarked"],drop_first=True) 
df=pd.get_dummies(df, columns=["deck"],drop_first=True) 
#binominally codes categorical variables
```

```python
df.isnull().any() 
```

```python
df=df.dropna() 
#missing values are dropped
```

```python
df
```

```python
initial_data=df
initial_data
#df is used a lot, for no errors initial dataset is duplicated
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array(df.drop(columns=['survived']))
Y = np.array(df['survived'].values)
#survived is set to be the dependent variable
```

```python
#kNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
```

```python
#cross validation to determine k
```

```python
cv = KFold(n_splits=10, random_state=1, shuffle=True)
```

```python
#for the first trial, splits = 10
```

```python
CVAccuracy=[]
for train_index, validation_index in cv.split(X):
    X_train, X_valid, Y_train, Y_valid = X[train_index], X[validation_index], Y[train_index], Y[validation_index]
    for j in range(1,26):
        knn = KNeighborsClassifier(n_neighbors = j)
        # Fit the classifier to the data
        knn.fit(X_train,Y_train)
        CVAccuracy.append([knn.score(X_valid, Y_valid),j])
    df = pd.DataFrame (CVAccuracy,columns=['Validation Accuracy','NeighbourSize'])
```

```python
#using a for loop, calculates validation score beginning from k=1 upto k=25
```

```python
kfoldCV = df.groupby("NeighbourSize")
kfoldCV = kfoldCV.mean()
kfoldCV = kfoldCV.reset_index()
kfoldCV[['NeighbourSize', 'Validation Accuracy']]
```

```python
TenfoldCVError = df.mean()
TenfoldCVError
```

```python
#mean validation accuracy is 0.693072 for 10 splits
```

```python
#to try the same precedure with 5 splits
```

```python
cv = KFold(n_splits=5, random_state=1, shuffle=True)
CVAccuracy=[]
for train_index, validation_index in cv.split(X):
    X_train, X_valid, Y_train, Y_valid = X[train_index], X[validation_index], Y[train_index], Y[validation_index]
    for j in range(1,26):
        knn = KNeighborsClassifier(n_neighbors = j)
        # Fit the classifier to the data
        knn.fit(X_train,Y_train)
        CVAccuracy.append([knn.score(X_valid, Y_valid),j])
    df = pd.DataFrame (CVAccuracy,columns=['Validation Accuracy','NeighbourSize'])
    kfoldCV = df.groupby("NeighbourSize")
kfoldCV = kfoldCV.mean()
kfoldCV = kfoldCV.reset_index()
kfoldCV[['NeighbourSize', 'Validation Accuracy']]
```

```python
FivefoldCVError = df.mean()
FivefoldCVError
```

```python
#mean validation accuracy is 0.693072 with 5 splits
```

```python
#splits=5 is chosen
#with 5 splits, best performance is at k=20 neighbours
```

```python
#trying the model with all the features
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1, stratify=Y)
```

```python
#train test split to measure model performance
```

```python
knn = KNeighborsClassifier(n_neighbors = 20)
knn.fit(X_train,Y_train)

knn.score(X_test, Y_test)
```

```python
#data is fit with training data and evaluated with test data
#score shows this algorithm classifies accurately 92.3% of the time
```

```python
Y_pred=knn.predict(X_test)
```

```python
from sklearn import metrics
cnf_matrix= metrics.confusion_matrix(Y_test,Y_pred)
cnf_matrix
```

```python
#nobody is falsely identified as survived
#11 people are falsely identified as dead which in this context might be a problem
```

```python
X=initial_data.drop("survived",axis=1)
y=initial_data["survived"].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1,stratify=y)
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
knn = KNeighborsClassifier(n_neighbors = 20)
sfs = SFS(knn, 
           k_features=15, 
           forward=True, 
           floating=False, 
           verbose=2,
           scoring='accuracy',
           cv=5)
feature_names = ("pclass", "sex","age","sibsp","parch","fare","embarked_Q","embarked_S","deck_B","deck_C","deck_D","deck_E","deck_F","deck_G")
sfs = sfs.fit(X_train, y_train, custom_feature_names=feature_names)
pd.DataFrame.from_dict(sfs.get_metric_dict()).T
```

```python
#sfs also provides same features as sbs
```

```python
#Best Subset Selection
```

```python
from mlxtend.feature_selection import ExhaustiveFeatureSelector as EFS
knn = KNeighborsClassifier(n_neighbors = 20)
efs = EFS(knn, 
           min_features=1,
           max_features=8,
           scoring='accuracy',
           print_progress=True,
           cv=5)

feature_names = ("pclass", "sex","age","sibsp","parch","fare","embarked_Q","embarked_S","deck_B","deck_C","deck_D","deck_E","deck_F","deck_G")
efs = efs.fit(X_train, y_train, custom_feature_names=feature_names)

print('Best accuracy score: %.2f' % efs.best_score_)
print('Best subset (indices):', efs.best_idx_)
print('Best subset (corresponding names):', efs.best_feature_names_)
```

```python
#unlike the sbs and sfs, efs recommended a model with 5 features
#those 4 features are pclass, sex, sibsp, deck_B, deck_E
```
